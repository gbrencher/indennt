```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# function to find downslope displacement for a single pixel
def downslope(los_d, slope, azimuth, heading=-170, incidence=35):
    slope = math.radians(slope)
    azimuth = math.radians(azimuth)
    heading = math.radians(heading)
    incidence = math.radians(incidence)
    a = heading - azimuth

    slope_d = los_d/(math.sin(a)*math.sin(incidence)*math.cos(slope)+math.cos(incidence)*math.sin(slope))

    return slope_d       
```

```python
downslope(1, slope=55, azimuth=75)
```

```python
azimuth_range = np.arange(0, 360, 1)
scalar_list = []
for azimuth in azimuth_range:
    scalar_list.append(downslope(1, 55, azimuth))

f, ax = plt.subplots()
ax.plot(azimuth_range, scalar_list)
```

```python
slope_range = np.arange(0, 90, 1)
scalar_list = []
for slope in slope_range:
    scalar_list.append(downslope(1, slope, 75))

f, ax = plt.subplots()
ax.plot(slope_range, scalar_list)
```
